Data Interpolation : Estimating unknown values

Linear interpolate

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
plt.scatter(x,y)
```

```python
x_new=np.linspace(1,5,10)
y_inter=np.interp(x_new,x,y)
plt.scatter(x_new,y_inter)
```

Cubic Interpolate

```python
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
f=interp1d(x,y,kind="cubic")
x_new=np.linspace(1,5,10)
y_interp=f(x_new)
plt.scatter(x_new,y_interp)
```

Polynomial Interpolate 

```python
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
# plt.scatter(x,y)
p=np.polyfit(x,y,4)
x_new=np.linspace(1,5,10)
y_interpolate=np.polyval(p,x_new)
plt.scatter(x_new,y_interpolate)
```
